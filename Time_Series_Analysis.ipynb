We will opt to recognize the range of 165K to 215K as **"Typical"** physical volume  

**Typical**: Physical volume within the range of 165K to 215K  
**Spike**: physical volume above the "typical" range  
**Dip**: Physical volume below the "typical" range  

### Tendencies   
* **Spikes** tend to occur about once a month between February and December
  * Likelihood and magnitude increases around major holidays
    * July 4, Labor Day, Thanksgiving, Christmas/New Year's
* **Dips** tend to occur in months following major holidays:
  * January, February, October
* Promotion prices may act as confounding factors on holiday spikes

#### -- Physical Volume + Promotion --  
Of the observed instances, the results imply an interesting narrative around the leading package type (**12OZ 12-Pack Cans**).
* This package type has been under extended periods of significant, uninterrupted promotion. This can complicate interpretations.
  * Constant promotion of this package type could influence its ranking among other package types
    * With limited promotion dollars, this is a potential scenario ripe for improvements
      * For example, diverting promotional dollars to **2L Bottles** may yield greater overall volume for a brand
  * The persistent promotion could be a cause and an effect on reduced promotions causing dips in volume
    * There is also possibly a scale factor involved.
    * A metric evaluating volume per promotion units could be valuable.
