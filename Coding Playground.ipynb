```
-- See price of an item at a store over time

select

s.pod_id

, s.product_key

, t.nielsen_period_date

, s.dollar_sales - s.any_price_decrease_dollar_sales as edv_dollar_sales

, s.unit_sales - s.any_price_decrease_unit_sales as edv_unit_sales

-- need to fill the null values with linear interpolation

, case when round(div0(edv_dollar_sales, edv_unit_sales),2) = 0

then null

else round(div0(edv_dollar_sales, edv_unit_sales),2)

end
as edv_price
, s.any_price_decrease_dollar_sales
, s.any_price_decrease_unit_sales
, round(div0(s.any_price_decrease_dollar_sales, s.any_price_decrease_unit_sales),2) as prom_price
, p.key_brand
, p.key_package
, p.item_desc
, l.store_name
, l.store_description

from

db_swire_bi_l_edw.transformed.fact_sales_nielsen_store_data s

inner join

db_swire_bi_l_edw.transformed.dim_sales_nielsen_store_product p
on s.product_key = p.product_key

inner join

db_swire_bi_l_edw.transformed.dim_sales_nielsen_store l
on s.pod_id = l.pod_id

inner join

db_swire_bi_l_edw.transformed.dim_sales_nielsen_period t
on s.period_key = t.period_key
where
p.route_to_market = 'BOTTLER COMP SET' -- only red truck
and p.key_trademark in ('KO TRADEMARK', 'DR PEPPER TRADEMARK') -- only coke products
and p.swire_total_category = 'TTL SSD' -- only sparkling for the poc
and l.nielsen_trade_channel_desc = 'Grocery - Supermarket-Conventional' -- only supermarket
and s.product_key = 41029250 -- Dr. Pepper Cherry 12 pack
--and s.product_key = 40982585 -- coke 12-pack
and s.pod_id = 149994 -- Safeway in Seattle
--and s.pod_id = 155456 -- Albertson's in Boise
order by
s.period_key desc
limit 100
```
